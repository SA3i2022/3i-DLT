```python
import os
import numpy as np
import tifffile as tiff

def create_directory(directory):
    if not os.path.exists(directory):
        os.makedirs(directory)

def save_subimages(tiff_data, output_dir, subimage_size=(256, 256)):
    depth, height, width = tiff_data.shape
    
    # Calculate the number of subimages in each dimension
    num_subimages_y = height // subimage_size[0]
    num_subimages_x = width // subimage_size[1]
    
    for z in range(depth):
        for y in range(num_subimages_y):
            for x in range(num_subimages_x):
                subimage = tiff_data[z, y*subimage_size[0]:(y+1)*subimage_size[0], 
                                        x*subimage_size[1]:(x+1)*subimage_size[1]]
                
                subimage_filename = f'subimage_z{z}_y{y}_x{x}.tif'
                tiff.imwrite(os.path.join(output_dir, subimage_filename), subimage)

def main(input_tiff_path, output_dir):
    # Create the output directory if it doesn't exist
    create_directory(output_dir)
    
    # Read the 3D TIFF file
    tiff_data = tiff.imread(input_tiff_path)
    
    # Check the dimensions
    if tiff_data.shape != (100, 800, 600):
        raise ValueError("The input TIFF file does not have the expected dimensions of 100x800x600.")
    
    # Save the subimages
    save_subimages(tiff_data, output_dir)

if __name__ == "__main__":
    input_tiff_path = r"C:\Users\colin\TrainingImages.tif"
    output_dir = r"C:\Users\colin\TrainingImages"
    
    main(input_tiff_path, output_dir)
```
